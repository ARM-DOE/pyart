```python
# import modules for use.
import pyart
import scipy
import numpy as np
import matplotlib.pyplot as plt
import plotly.plotly as py
import plotly.graph_objs as go

from mpl_toolkits.mplot3d import Axes3D
from plotly.graph_objs import Surface, Layout, Scene
from plotly.offline import(
    download_plotlyjs, init_notebook_mode, plot, iplot)

init_notebook_mode()
%matplotlib inline
```

```python
# reading a nexrad file with pyart to convert into a radar object.
filename = '/home/zsherman/training_exercises/data/KLOT20130417_235520_V06.gz'
myradar = pyart.io.read(filename)
```

```python
# Use py-art's new vad function.
# Function creates dict with u_mean, v_mean, and z_interval.
vad = pyart.retrieve.velocity_azimuth_display(myradar,
                                              velocity='velocity',
                                              z_count=101)
```

```python
# Simple plot to show u_mean and v_mean as height increases.
fig = plt.figure()
plt.subplot(1, 2, 1)
plt.plot(vad.u_wind, vad.height / 1000.0, 'b-')
plt.plot(vad.v_wind, vad.height / 1000.0, 'r-')
plt.xlabel(r'Velocity (ms$^{-1}$)')
plt.ylabel(r'Height (km)')
plt.title('VAD, u=blue, v=red')
#plt.legend()
plt.show()
```

```python
# To get fancier, here is an example on how to generate a 3D model.
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
x1 = vad.u_wind
x2 = vad.v_wind
y = vad.height / 1000
plt.xlim(-12, 12)
plt.ylim(-12, 12)
ax.title.set_position([.5, 1.05])
ax.set_title('Velocity Azimuth Display (VAD)')
ax.set_xlabel('Velocity (m/s)')
ax.set_ylabel('Velocity (m/s)')
ax.set_zlabel('Height (km)')
ax.plot(x1, np.zeros_like(x1), y, label='U Mean')
ax.plot(np.zeros_like(x2), x2, y, '-r', label='V Mean')
# default ax.azim = 300, default ax.elev = 30
ax.legend()
plt.show()
```

```python
# The elevation and azimuth can be adjusted for different views.
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
plt.xlim(-12, 12)
plt.ylim(-12, 12)
ax.title.set_position([.5, 1.05])
ax.set_title('Velocity Azimuth Display (VAD)')
ax.set_xlabel('Velocity (m/s)')
ax.set_ylabel('Velocity (m/s)')
ax.set_zlabel('Height (km)')
ax.plot(x1, np.zeros_like(x1), y, label='U Mean')
ax.plot(np.zeros_like(x2), x2, y, '-r', label='V Mean')
ax.azim = 45
ax.elev = 20
ax.legend()
plt.show()
```

```python
# We can even take VAD a make a nice 3d interactive plot.
# Requires 'pip install plotly' to use

x1 = vad.u_wind
x2 = vad.v_wind
z = vad.height / 1000

trace1 = go.Scatter3d(
    x=x1, y=np.zeros_like(x1), z=z,
    name='U Mean',
    marker=dict(
        size=2,
        color='#0000FF'),
    line=dict(
        color='#0000FF',
        width=1))
trace2 = go.Scatter3d(
    x=np.zeros_like(x2), y=x2, z=z,
    name='V Mean',
    marker=dict(
        size=2,
        color='#FF0000'),
    line=dict(
        color='#FF0000',
        width=1))

layout = dict(
    width=900,
    height=700,
    autosize=False,
    title='Velocity Azimuth Display (VAD)',
    scene=dict(
        xaxis=dict(
            range=[-12, 12],
            title='Velocity (m/s)',
            gridcolor='rgb(255, 255, 255)',
            zerolinecolor='rgb(255, 255, 255)',
            showbackground=True,
            backgroundcolor='rgb(230, 230, 230)'),
        yaxis=dict(
            range=[-12, 12],
            title='Velocity (m/s)',
            gridcolor='rgb(255, 255, 255)',
            zerolinecolor='rgb(255, 255, 255)',
            showbackground=True,
            backgroundcolor='rgb(230, 230, 230)'),
        zaxis=dict(
            title='Height (km)',
            gridcolor='rgb(255, 255, 255)',
            zerolinecolor='rgb(255, 255, 255)',
            showbackground=True,
            backgroundcolor='rgb(230, 230, 230)')))

data = [trace1, trace2]
fig = dict(data=data, layout=layout)
iplot(fig)
```
